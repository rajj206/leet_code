```python
def countTriplets(arr, r):
    from collections import defaultdict
    
    potential_pairs = defaultdict(int)  # This keeps track of potential second elements of pairs
    potential_triplets = defaultdict(int)  # This keeps track of potential third elements of triplets
    triplet_count = 0  # This counts the total number of triplets found
    
    for value in arr:
        # Step 1: Check if current value can complete any triplets
        if value in potential_triplets:
            triplet_count += potential_triplets[value]
        
        # Step 2: Check if current value can be the second element in any triplet
        if value in potential_pairs:
            potential_triplets[value * r] += potential_pairs[value]
        
        # Step 3: Assume current value is the start of a new pair
        potential_pairs[value * r] += 1
    
    return triplet_count

# Example usage:
if __name__ == "__main__":
    n = 4
    r = 2
    arr = [1, 2, 2, 4]
    print(countTriplets(arr, r))  # Expected Output: 2
```
