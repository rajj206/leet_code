With brute force appraoch its going to be n3 . Better way is below by starting from center

```python
def palindrome(s):
    res = ""
    resLen = 0
    for i in range(len(s)):
        # Odd length
        l,r = i,i 
        print (l)
        print (r)
        while l>=0 and r < len(s) and s[l] == s[r]:
            if (r-l+1) > resLen:
                res=s[l:r+1]
                resLen = r-l+1
            l-=1
            r+=1
        # Even Length
        l,r=i,i+1
        while l>=0 and r < len(s) and s[l] == s[r]:
            if (r-l+1) > resLen:
                res=s[l:r+1]
                resLen = r-l+1
            l-=1
            r+=1
    return res            
```
