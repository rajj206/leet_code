Here's the approach:

1) Count the 'a's in the given string s: 
> Determine how many times 'a' appears in s.

2) Determine the number of complete repeats of s within the first n characters:
> Calculate how many full repetitions of s fit into n characters.
> This can be done using integer division n//len(s).

3) Determine the number of 'a's contributed by these full repetitions:
> Multiply the count of 'a's in s by the number of full repetitions.

4) Handle the remaining characters:
> Determine the remaining characters after accounting for the full repetitions.
> Count the 'a's in this leftover substring.

5) Sum the 'a's from the full repetitions and the leftover part:
> Add the number of 'a's from the full repetitions to the number of 'a's in the leftover part to get the total.

```python
def repeatedString(s, n):
    # Count the number of 'a's in the original string
    a_count_in_s = s.count('a')
    
    # Calculate the number of full repetitions of s within n characters
    full_repeats = n // len(s)
    
    # Calculate the number of 'a's in these full repetitions
    total_a_count = full_repeats * a_count_in_s
    
    # Calculate the remaining characters after the full repetitions
    remainder = n % len(s)
    
    # Count the number of 'a's in the remaining substring
    total_a_count += s[:remainder].count('a')
    
    return total_a_count


```

```python
# Sample Input
s = "aba"
n = 10

# Sample Output
print(repeatedString(s, n))  # Output should be 7
```
