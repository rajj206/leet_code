To solve the problem of finding how many pairs of socks with matching colors there are in a given pile, we can use a dictionary to count the occurrences of each sock color. 
Once we have the counts, we can easily determine the number of pairs for each color.

Detailed Steps:
1) Count the Occurrences:
> Use a dictionary to count how many times each sock color appears in the array.

2) Calculate Pairs:
> For each sock color, the number of pairs is given by the integer division of the count by 2.

```python
def sockspair(n, ar):
    from collections import Counter

    color_counts = Counter(ar)
    print (color_counts)

    pairs = 0
    for count in color_counts.values():
        pairs+=count//2

    return pairs
```

```python
n = 9
ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]
print(sockspair(n, ar))  # Expected Output: 3
```
