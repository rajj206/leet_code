```python
def lengthOfLongestSubstring(s: str) -> int:
    char_set = set()
    left = 0
    max_length = 0
    
    for right in range(len(s)):
        # If the character at 'right' is in the set, move 'left' pointer to the right
        print ('before right : ' + str(right))
        while s[right] in char_set:
            char_set.remove(s[left])
            print ('in loop char_set : ' + str(char_set))
            left += 1
            print ('left :' + str(left))
        
        # Add the character at 'right' to the set
        char_set.add(s[right])
        print ('out loop char_set : ' + str(char_set))
        print ('after right : ' + str(right))
        
        # Update max_length if the current window is larger
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Example usage:
print(lengthOfLongestSubstring("abcdaabcadef"))  # Output: 3
```
